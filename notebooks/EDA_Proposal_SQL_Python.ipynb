```python
%%sql
SELECT * AS column_name, type AS declared_type, "notnull", dflt_value
FROM pragma_table_info('statewise');
```

```python
%%sql
SELECT *
FROM statewise
LIMIT 5;
```

```python
%%sql
SELECT
  COUNT(*) AS n_rows,
  SUM(CASE WHEN NULLIF(state,'') IS NULL THEN 1 ELSE 0 END) AS n_blank_state
FROM statewise;
```

```python
%%sql
SELECT
  COUNT(*) AS n_rows,
  COUNT(DISTINCT state) AS n_states
FROM statewise;
```

```python
%%sql
SELECT
  'population' AS col, SUM(CASE WHEN NULLIF(population,'') IS NULL THEN 1 ELSE 0 END) AS n_null
UNION ALL
SELECT 'infant_mortality', SUM(CASE WHEN NULLIF(infant_mortality,'') IS NULL THEN 1 ELSE 0 END)
UNION ALL
SELECT 'under5_mortality', SUM(CASE WHEN NULLIF(under5_mortality,'') IS NULL THEN 1 ELSE 0 END)
ORDER BY n_null DESC;
```
