```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Features with prefixes
PREFIXES = [
    "YY_Infant_Mortality_Rate",
    "YY_Under_Five_Mortality_Rate",
    "YY_Neo_Natal_Mortality_Rate",
    "YY_Post_Neo_Natal_Mortality_Rate",
]
# Candidate grouping columns (used for "Top N groups" plots)
GROUP_COL_CANDIDATES = ["State", "State_District_Name"]
def find_group_col(df: pd.DataFrame) -> str | None:
    """Return the first grouping column found in df, else None."""
    for c in GROUP_COL_CANDIDATES:
        if c in df.columns:
            return c
    return None


def cols_with_prefixes(df: pd.DataFrame, prefixes: list[str]) -> list[str]:
    """Return all columns that start with any of the given prefixes (deduped, order preserved)."""
    cols: list[str] = []
    for p in prefixes:
        cols.extend([c for c in df.columns if str(c).startswith(p)])
    # Deduplicate while preserving order
    return list(dict.fromkeys(cols))


def _clean_label(colname: str) -> str:
    """Remove a single leading 'YY_' from a column label for readability."""
    return str(colname)[3:] if str(colname).startswith("YY_") else str(colname)


def melt_features(df: pd.DataFrame, prefixes: list[str]) -> pd.DataFrame:
    """
    Melt wide 'YY_*' mortality columns to tidy format.
    Adds:
      - feature_full  : original column name
      - feature_clean : column name with leading 'YY_' removed (if present)
    """
    yy_cols = cols_with_prefixes(df, prefixes)
    if not yy_cols:
        raise ValueError("No columns found for the provided PREFIXES. Check names and casing.")

    id_vars = [c for c in df.columns if c not in yy_cols]
    tidy = df.melt(id_vars=id_vars, value_vars=yy_cols,
                   var_name="yy_feature", value_name="value")
    tidy["value"] = pd.to_numeric(tidy["value"], errors="coerce")
    tidy["feature_full"] = tidy["yy_feature"]
    tidy["feature_clean"] = tidy["yy_feature"].map(_clean_label)
    return tidy

# Data Plots
def plot_overall_means_by_feature(tidy: pd.DataFrame, use_clean_labels: bool = True) -> None:
    """
    Bar chart of overall means per feature.
    Set use_clean_labels=False to use full original names.
    """
    label_col = "feature_clean" if use_clean_labels else "feature_full"
    agg = (tidy.groupby(label_col, dropna=True)["value"]
           .mean()
           .reset_index()
           .sort_values("value", ascending=False))

    plt.figure(figsize=(10, 5))
    plt.bar(agg[label_col], agg["value"])
    plt.title("Overall Mean by Feature")
    plt.ylabel("Mean Value")
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()
    plt.show()


def plot_top_groups_for_feature(
    tidy: pd.DataFrame,
    feature_label: str,
    group_col: str | None,
    top_n: int = 10,
    use_clean_labels: bool = True
) -> None:
    """
    Horizontal bar chart of top-N groups (e.g., States/Districts) by mean for a single feature.
    """
    if group_col is None:
        print(f"[info] No group column found; skipping group ranking for '{feature_label}'.")
        return

    label_col = "feature_clean" if use_clean_labels else "feature_full"
    sub = tidy.loc[tidy[label_col] == feature_label].copy()
    if sub.empty:
        print(f"[warn] Feature '{feature_label}' not found in tidy data (label_col={label_col}).")
        return

    agg = (sub.groupby(group_col, dropna=True)["value"]
           .mean()
           .sort_values(ascending=False)
           .head(top_n))

    plt.figure(figsize=(9, 6))
    plt.barh(agg.index[::-1], agg.values[::-1])
    plt.title(f"Top {top_n} {group_col}s by Mean {feature_label}")
    plt.xlabel("Mean Value")
    plt.tight_layout()
    plt.show()


def plot_corr_of_selected(df: pd.DataFrame, prefixes: list[str], use_clean_labels: bool = True) -> None:
    """
    Heatmap of correlations among selected YY_* columns.
    Axis tick labels can use cleaned names.
    """
    sel_cols = cols_with_prefixes(df, prefixes)
    if not sel_cols:
        print("[warn] No columns found for the given prefixes.")
        return

    # Numeric conversion
    num = df[sel_cols].apply(pd.to_numeric, errors="coerce")

    # Compute correlation
    corr = num.corr()

    # Build labels
    labels = [_clean_label(c) for c in sel_cols] if use_clean_labels else sel_cols

    plt.figure(figsize=(10, 8))
    plt.imshow(corr.values, interpolation="nearest")
    plt.colorbar()
    plt.title("Correlation (Selected Mortality Features)")
    plt.xticks(range(len(labels)), labels, rotation=90)
    plt.yticks(range(len(labels)), labels)
    plt.tight_layout()
    plt.show()

def run_prelim_eda(df: pd.DataFrame, prefixes: list[str] = None, top_n: int = 10) -> None:
    """
    Execute a concise EDA flow:
      - Melt to tidy format with full & clean labels
      - Print feature inventory
      - Plot overall means
      - Plot top-N groups per feature
      - (Optional) Correlation heatmap
    """
    if prefixes is None:
        prefixes = PREFIXES

    group_col = find_group_col(df)
    tidy = melt_features(df, prefixes)

    # Inventory
    feature_list = tidy["feature_clean"].unique().tolist()
    print("Health Analytics Data Overview")
    print("Features Column Data (Per State):")
    for f in feature_list:
        print(" -", f)
    print("\n[INFO] Feature Column:", group_col)

    # Overall means
    plot_overall_means_by_feature(tidy, use_clean_labels=True)

    # Top-N per feature
    for f in sorted(tidy["feature_clean"].dropna().unique()):
        plot_top_groups_for_feature(tidy, f, group_col, top_n=top_n, use_clean_labels=True)

    # Correlation among selected features (toggle if desired)
    plot_corr_of_selected(df, prefixes, use_clean_labels=True)


# -------------------------------
# Run EDA pipeline:
# -------------------------------
group_col = find_group_col(df)
tidy = melt_features(df, PREFIXES)
run_prelim_eda(df, PREFIXES, top_n=10)

# # Full original labels for a specific plot:
# plot_overall_means_by_feature(tidy, use_clean_labels=False)
# plot_top_groups_for_feature(tidy, "Under_Five_Mortality_Rate", group_col, top_n=10, use_clean_labels=True)
# plot_corr_of_selected(df, PREFIXES, use_clean_labels=True)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# --- Check feature columns ---
PREFIXES = ["YY_Infant_Mortality_Rate", "YY_Under_Five_Mortality_Rate", "YY_Neo_Natal_Mortality_Rate", "YY_Post_Neo_Natal_Mortality_Rate"] 
GROUP_COL_CANDIDATES = ["State_Name", "State_District_Name"] # can add more feature columns here (by group/category. state, locations, etc)

# --- group columns ---
def find_group_col(df):
    for c in GROUP_COL_CANDIDATES:
        if c in df.columns:
            return c
    return None  

def cols_with_prefixes(df, prefixes):
    cols = []
    for p in prefixes:
        cols.extend([c for c in df.columns if c.startswith(p)])
    return list(dict.fromkeys(cols))  

def melt_prefixes(df, prefixes):
    yy_cols = cols_with_prefixes(df, prefixes)
    id_vars = [c for c in df.columns if c not in yy_cols]
    tidy = df.melt(id_vars=id_vars, value_vars=yy_cols,
                   var_name="yy_feature", value_name="value")
    tidy["value"] = pd.to_numeric(tidy["value"], errors="coerce")
    # add top-level prefix label for quick faceting/grouping
    tidy["prefix"] = tidy["yy_feature"].str.extract(r"^(YY_[A-Za-z]+)")
    return tidy

def plot_overall_means_by_prefix(tidy):
    agg = tidy.groupby("prefix", dropna=True)["value"].mean().reset_index()
    plt.figure(figsize=(7,4))
    plt.bar(agg["prefix"], agg["value"])
    plt.title("Overall Mean by Prefix Feature Column")
    plt.ylabel("Mean Value")
    plt.xticks(rotation=15, ha="right")
    plt.tight_layout()
    plt.show()

def plot_top_groups_for_prefix(tidy, prefix, group_col, top_n=10):
    if group_col is None:
        print(f"[info] No group column found; skipping group ranking for {prefix}.")
        return
    sub = tidy[(tidy["prefix"] == prefix)].copy()
    agg = sub.groupby(group_col, dropna=True)["value"].mean().sort_values(ascending=False).head(top_n)
    plt.figure(figsize=(8,5))
    plt.barh(agg.index[::-1], agg.values[::-1])
    plt.title(f"Top {top_n} {group_col}s by Mean {prefix}")
    plt.xlabel("Mean Value")
    plt.tight_layout()
    plt.show()

def plot_corr_of_selected(df, prefixes):
    sel_cols = cols_with_prefixes(df, prefixes)
    if not sel_cols:
        print("[warn] No columns found for the given prefixes.")
        return
    num = df[sel_cols].apply(pd.to_numeric, errors="coerce")
    corr = num.corr()
    plt.figure(figsize=(8,6))
    plt.imshow(corr, interpolation="nearest")
    plt.colorbar()
    plt.title("Correlation (selected YY_* columns)")
    plt.xticks(range(len(sel_cols)), sel_cols, rotation=90)
    plt.yticks(range(len(sel_cols)), sel_cols)
    plt.tight_layout()
    plt.show()

# --- Preliminary on Health Analytics Data ---
# Check dataframes (By State - preliminary EDA ond data)
group_col = find_group_col(df)
tidy = melt_prefixes(df, PREFIXES)

# Check column data per prefix
counts = (pd.Series(cols_with_prefixes(df, PREFIXES))
          .str.extract(r"^(YY_[A-Za-z]+)")[0]
          .value_counts())
print("Health Analytics Data Overview:")
print("Features (By State):\n", counts.to_string(), "\n")

# Overall average means by prefix
plot_overall_means_by_prefix(tidy)

# Highest feature column numbers (e.g., States/Districts) for each prefix
for p in sorted(tidy["prefix"].dropna().unique()):
    plot_top_groups_for_prefix(tidy, p, group_col, top_n=10)

# Correlation among all selected YY_* column data
#plot_corr_of_selected(df, PREFIXES)
```
