```python
import pandas as pd

# List of prefixes to search
prefixes = ["YY_Under", "YY_Neo", "YY_Post", "YY_Infant"]

# Function to filter column names by prefixes
def get_matching_columns(df, prefixes):
    return [col for col in df.columns if any(pref in col for pref in prefixes)]

# Extract matches
df_cols = get_matching_columns(df, prefixes)
df2_cols = get_matching_columns(df2, prefixes)

# Convert to DataFrames for stdout
df_cols_df = pd.DataFrame(df_cols, columns=["df Columns"])
df2_cols_df = pd.DataFrame(df2_cols, columns=["df2 Columns"])

print("Preliminary EDA for Infant Mortality Rates. Columns:")
print("Matching columns in df:")
print(df_cols_df)

print("\nMatching columns in df2:")
print(df2_cols_df)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# --- config ---
PREFIXES = ["YY_Infant", "YY_Under", "YY_Neo", "YY_Post"]  
GROUP_COL_CANDIDATES = ["State_Name", "State_District_Name"] 

# --- helpers ---
def find_group_col(df):
    for c in GROUP_COL_CANDIDATES:
        if c in df.columns:
            return c
    return None  

def cols_with_prefixes(df, prefixes):
    cols = []
    for p in prefixes:
        cols.extend([c for c in df.columns if c.startswith(p)])
    return list(dict.fromkeys(cols))  

def melt_prefixes(df, prefixes):
    yy_cols = cols_with_prefixes(df, prefixes)
    id_vars = [c for c in df.columns if c not in yy_cols]
    tidy = df.melt(id_vars=id_vars, value_vars=yy_cols,
                   var_name="yy_feature", value_name="value")
    # coerce numeric safely
    tidy["value"] = pd.to_numeric(tidy["value"], errors="coerce")
    # add top-level prefix label for quick faceting/grouping
    tidy["prefix"] = tidy["yy_feature"].str.extract(r"^(YY_[A-Za-z]+)")
    return tidy

def plot_overall_means_by_prefix(tidy):
    agg = tidy.groupby("prefix", dropna=True)["value"].mean().reset_index()
    plt.figure(figsize=(7,4))
    plt.bar(agg["prefix"], agg["value"])
    plt.title("Overall mean by prefix")
    plt.ylabel("Mean value")
    plt.xticks(rotation=15, ha="right")
    plt.tight_layout()
    plt.show()

def plot_top_groups_for_prefix(tidy, prefix, group_col, top_n=10):
    if group_col is None:
        print(f"[info] No group column found; skipping group ranking for {prefix}.")
        return
    sub = tidy[(tidy["prefix"] == prefix)].copy()
    agg = sub.groupby(group_col, dropna=True)["value"].mean().sort_values(ascending=False).head(top_n)
    plt.figure(figsize=(8,5))
    plt.barh(agg.index[::-1], agg.values[::-1])
    plt.title(f"Top {top_n} {group_col}s by mean {prefix}")
    plt.xlabel("Mean value")
    plt.tight_layout()
    plt.show()

def plot_corr_of_selected(df, prefixes):
    sel_cols = cols_with_prefixes(df, prefixes)
    if not sel_cols:
        print("[warn] No columns found for the given prefixes.")
        return
    num = df[sel_cols].apply(pd.to_numeric, errors="coerce")
    corr = num.corr()
    plt.figure(figsize=(8,6))
    plt.imshow(corr, interpolation="nearest")
    plt.colorbar()
    plt.title("Correlation (selected YY_* columns)")
    plt.xticks(range(len(sel_cols)), sel_cols, rotation=90)
    plt.yticks(range(len(sel_cols)), sel_cols)
    plt.tight_layout()
    plt.show()

# --- run ---
# Check dataframes (By State - preliminary EDA ond data)
group_col = find_group_col(df)
tidy = melt_prefixes(df, PREFIXES)

# 1) Check columns per prefix)
counts = (pd.Series(cols_with_prefixes(df, PREFIXES))
          .str.extract(r"^(YY_[A-Za-z]+)")[0]
          .value_counts())
print("Columns found per prefix:\n", counts.to_string(), "\n")

# 2) Overall means by prefix (bar chart)
plot_overall_means_by_prefix(tidy)

# 3) Top groups (e.g., States/Districts) for each prefix
for p in sorted(tidy["prefix"].dropna().unique()):
    plot_top_groups_for_prefix(tidy, p, group_col, top_n=10)

# 4) Correlation heatmap among all selected YY_* columns
#plot_corr_of_selected(df, PREFIXES)
```
